```python
import json
from pathlib import Path
```

```python
files = Path.cwd().glob('*.json')
for f in files:
    print(f)
    with open(f, "r") as file:
        data = json.load(file)
    data['zeta'] = [ z/v for z,v in zip(data['zeta'],data['rvv']) ]
    with open(f, "w") as file:
        json.dump(data,file)
```
